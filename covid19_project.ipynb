```python
covid_df['Date']=pd.to_datetime(covid_df['Date'])
```

## Find Active cases so Add new Column

```python
covid_df['Active_cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.isnull().sum()
```

```python
covid_df.dropna(inplace=True)
```

```python
covid_df.tail()
```

```python
statewise=pd.pivot_table(covid_df,values=['Cured','Deaths','Confirmed'],index='State/UnionTerritory',aggfunc=max)
```

```python
statewise['Recover_rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background.gradient(cmap='cubehlix')
```

## Find out Top 10 most active Cases States in india

```python
covid_df.head()
```

```python
top_10_state=covid_df.groupby('State/UnionTerritory').max()[['Active_cases']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
plt.figure(figsize=(15,5))
plt.title('Top 10 Active Cases States in india')
plt.xlabel('Top 10 States')
plt.ylabel('Active Cases')
sns.barplot(x='State/UnionTerritory',y='Active_cases',data=top_10_state.iloc[:10],linewidth=2)
plt.show()
```

## Find Out Top States with Highest deaths

```python
statewise.head()
```

```python
States_M=statewise.groupby('State/UnionTerritory').max()[['Deaths']].sort_values(by=['Deaths'],ascending=False).reset_index()
```

```python
plt.figure(figsize=(15,5))
plt.xlabel('Top 10 States',size=15)
plt.ylabel('Total death cases',size=15)
plt.title('Top 10 States with Highest deaths',size=20 )
sns.barplot(x='State/UnionTerritory',y='Deaths',data=States_M.iloc[:10],linewidth=2)
plt.show()
```

## Find the Top State vacinated in India?

```python
vac_df=vac_df[vac_df['State']!='India']
```

```python
vac_df.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
```

```python
state_high_vac=vac_df.groupby('State')['Total'].sum().to_frame('Total')
state_high_vac=state_high_vac.sort_values(by='Total',ascending=False)[:5]
state_high_vac
```

```python
plt.figure(figsize=(10,4))
plt.title('Top Five Vaccinated States',size=15)
plt.xlabel('States',size=14)
plt.ylabel('Total vaccination',size=14)
sns.barplot(data=state_high_vac,x='State',y='Total',edgecolor='black',palette=['Blue','Orange','Green','Red','Purple'])
plt.show()
```

## Find the State with least number of vaccination indiviual ?

```python
state_no_vac=vac_df.groupby('State')['Total'].sum().to_frame('Total')
state_no_vac=state_no_vac.sort_values(by='Total',ascending=True)[:5]
```

```python
plt.figure(figsize=(15,4))
plt.title('Bottom Five vacinated in india',size=15)
plt.xlabel('States',size=14)
plt.ylabel('Vaccinated',size=14)
sns.barplot(data=state_no_vac,x='State',y='Total',edgecolor='black',palette=['Blue','Orange','Green','Red','Purple'])
plt.show()
```
